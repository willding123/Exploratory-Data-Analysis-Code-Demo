```python
# pandas and numpy are the basic data science packages on Python 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
sns.set(color_codes = True) # use seaborn palette
```

```python
data = pd.read_csv("investments_VC.csv")
```

```python
# return first 10 rows of data

data.head(10)
#data.tail(5)
```

```python
# summary of data 
data.describe()
```

```python
#checking column data types 
data.dtypes
```

```python
#checking columns 
data.columns
```

```python
# renaming columns 
data =  data.rename(columns = {" funding_total_usd ":"funding_total_usd"," market " :"market"})
```

```python
# select columns 
df = data[["name","market","city","country_code","funding_total_usd","founded_year","status","funding_rounds"]]
```

```python
# print out the number of rows and columns 
df.shape
```

```python
# check duplicated rows
duplicates = df[df.duplicated()]
print("number of duplicate rows: ", duplicates.shape)
```

```python
# return number of data for each variable 
df.count()
```

```python
# drop duplicates 
df = df.drop_duplicates()
```

```python
# check the count after dropping duplicates 
df.count()
```

```python
# Find the number of nulls for each column.
print(df.isnull().sum())
```

```python
#drop nulls and na 
df = df.dropna() 
df.count()
```

```python
print(df.isnull().sum())
```

```python
# check column names 
df.columns
df["funding_total_usd"]
```

```python
# drop rows with non-numeric values 
indexname= df[df['funding_total_usd'] ==" -   " ].index
df.drop(indexname, inplace = True)


```

```python
# turn object into string then to float
#df["funding_total_usd"] = df["funding_total_usd"].str.replace(",","").astype(str).astype(float)

df.describe()
```

```python
# heat map
plt.figure(figsize=(20,10))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
sns.boxplot(x=df["funding_total_usd"])
```

```python
sns.boxplot(x=df["funding_rounds"])

```

```python
# number of unique values in market
df["market"].nunique()

df.groupby("market").sum().sort_values(by="funding_total_usd",ascending = False).head(5)

```
